```python
import warnings
warnings.filterwarnings("ignore")

import cv2
from fer import FER
from google.colab.patches import cv2_imshow

img = cv2.imread('/content/Sad Face 2.jpg')
detector = FER()
results = detector.detect_emotions(img)
emotion, score = detector.top_emotion(img)

x,y,w,h = results[0]['box']
cv2.rectangle(img, (x,y), (x+w, y+h),(123,0,123),2)
cv2_imshow(img)

print(emotion, score)
```
